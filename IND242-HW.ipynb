```python
#2.b:
import numpy as np
import pandas as pd

data = np.array([514, 476, 497, 511, 484, 513, 471, 470, 441, 466, 443, 481, 502, 528, 459, 548, 521, 517, 463, 478, 473, 514, 542, 519, 522, 523, 546, 487, 486, 473, 527, 470, 440, 564, 499, 523, 484, 463, 461, 437, 555, 525, 461, 539, 466, 470, 486, 490, 543, 519])

class_size = 16

freq_table = pd.DataFrame(columns=['Sınıf', 'Frekans'])

for i in range(8):
    lower_limit = min(data) + i * class_size
    upper_limit = lower_limit + class_size
    freq = ((data >= lower_limit) & (data < upper_limit)).sum()
    freq_table.loc[i] = f"{lower_limit} - {upper_limit}", freq

freq_table    
```

```python
#2.c: En yüksek frekansa sahip sınıf aralığı, 469-485 aralığıdır.
```

```python
#2.d:
total_freq = freq_table['Frekans'].sum()
freq_table['Göreli Frekans'] = (freq_table['Frekans'] / total_freq)
freq_table['Yüzdelik Frekans'] = (freq_table['Frekans'] / total_freq) * 100    
freq_table
```

```python
#2.e:
import matplotlib.pyplot as plt

limits=[437,453,469,485,501,517,533,549,565]

plt.figure(figsize = (15,10))
plt.hist(data, bins=8, edgecolor='black')
plt.xlabel('Sınıflar')
plt.xticks(limits)
plt.ylabel('Yüzde Frekanslar')
plt.title('Yüzde Frekans Dağılımı')
plt.show()
```

```python
#2.f: En çok 469-485 aralığında değerler görüyoruz, çok dağınık bir veri dizisi değil ama sağdan çarpık diyebiliriz.
```
